```python
def merge_sort(A):
  size = len(A)

  if size <= 1:
    return A

  mid = size//2
  s1 = merge_sort(A[0:mid])
  s2 = merge_sort(A[mid:])

  i = j = 0
  m, n = len(s1), len(s2)
  temp = []
  while True:
    if i == m:
      temp.extend(s2[j:])
      break
    elif j == n:
      temp.extend(s1[i:])
      break
    elif s1[i] < s2[j]:
      temp.append(s1[i])
      i+=1
    elif s2[j] < s1[i]:
      temp.append(s2[j])
      j+=1

  return temp
```

```python
arr = [5,2,6,1,9,7,0,8]
merge_sort(arr)
```
